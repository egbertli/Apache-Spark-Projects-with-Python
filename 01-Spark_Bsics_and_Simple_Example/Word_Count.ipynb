# MAPS VS. FLATMAP

###### Source data:
- The quick red
- fox jumped
- over the lazy
- brown dogs

## MAP() TRANSFORMS EACH ELEMENT OF RDD INTO ONE NEW ELEMENT

```python
lines = sc.text('redfox.txt')
```
```python
rageCaps = lines.map(lambda x: x.upper())
```

```
THE QUICK RED 
FOX JUMPED 
OVER THE LAZY 
BROWN DOGS**
```

## FLATMAP() CAN CREATE MANY NEW ELEMENTS FROM EACH ONE

```python
lines = sc.text('redfox.txt')
```
```python
words = lines.map(lambda x: x.split())
```

```
The
quick
red
fox
jumped
over
the
lazy
brown
dogs
```

````python
from pyspark import SparkConf, SparkContext
````

````python
conf = SparkConf().setMaster('local').setAppName('WordCount')
sc = SparkContext(conf=conf)
````

````python
books = sc.textFile('Book.txt')
words = books.flatMap(lambda line: line.split())
word_counts = words.countByValue()
````

````python
for word, count in word_counts.items():
    clean_word = word.encode('ascii', 'ignore')
    if clean_word:
        print(clean_word, " : ", count)
````

# Improving performance

````python
import re
def normalizeWords(text):
    return re.compile(r'\W+', re.UNICODE).split(text.lower())
    # return re.split(r'\W+', text)
````

````python
words = books.flatMap(normalizeWords)
word_count = words.countByValue()
````

````python
for word, count in word_count.items():
    clean_word = word.encode('ascii', 'ignore')
    if clean_word:
        print(clean_word, ' ', count)
````

# Sorting the results

###### STEP 1
```
- we cound sort what countByValue() returns, but let's use RDD's to keep it scalable.
- First, do countByValue() the 'hard way':
```
```python
wordcounts = words.map(lambda x:(x,1)).reduceByKey(lambda x, y: x + y)
```

```
- Convert each word to a key/value pair with a value of 1
- Then count them all up with reduceByKey()
```

###### STEP 2
```
- Flip the (word, count) pairs to (count, word)
- Then use sortByKey() to sort by count (now that count is our new key)
```
```python
wordcounts_sorted = word_count.map(lambda (x,y): (y,x))).sortByKey()
```

````python
word_counts = words.map(lambda word:(word, 1)).reduceByKey(lambda x,y: x + y)
type(word_counts)
````

````python
word_counts_sorted = word_counts.sortBy(lambda x: x[1], ascending=False)
````

````python
results = word_counts_sorted.collect()
````

````python
for result in results:
    word = result[0].encode('ascii', 'ignore')
    if word:
        print(word, " ", result[1])
````
